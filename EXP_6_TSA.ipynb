```python
# prompt: . Implement program to apply moving average smoothing for data preparation and time series forecasting the output as graph plot before applying moving average smoothing in datas and after results as graph plot

import pandas as pd
import matplotlib.pyplot as plt

def moving_average_smoothing(data, window_size):
    """
    Applies moving average smoothing to a time series.

    Args:
        data: A pandas Series representing the time series data.
        window_size: The size of the moving average window.

    Returns:
        A pandas Series representing the smoothed time series data.
    """
    # Calculate the moving average
    smoothed_data = data.rolling(window=window_size, center=True).mean()
    return smoothed_data

# Sample time series data (replace with your actual data)
data = pd.Series([1, 5, 2, 8, 3, 9, 4, 7, 6, 10])
# Create index for time series
data.index = pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05',
                              '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09', '2024-01-10'])


window_size = 3  # Adjust the window size as needed

# Apply moving average smoothing
smoothed_data = moving_average_smoothing(data, window_size)


# Plot the original and smoothed data
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data', marker='o')
plt.plot(smoothed_data, label=f'{window_size}-Day Moving Average', marker='x', linestyle='--')
plt.xlabel('Date')  # Corrected x-axis label
plt.ylabel('Value')
plt.title('Moving Average Smoothing')
plt.legend()
plt.grid(True)
plt.show()
```
